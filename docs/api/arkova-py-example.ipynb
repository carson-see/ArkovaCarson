# Arkova Python SDK example

Use the typed Python SDK to search records, verify a fingerprint, and inspect organization context.

```python
import os
from arkova import Arkova

arkova = Arkova(api_key=os.environ["ARKOVA_API_KEY"])
results = arkova.search("registered nurse", type="record", limit=5)
results.results[:2]
```

```python
fingerprint = "a" * 64
verification = arkova.verify_fingerprint(fingerprint)
verification.model_dump()
```

```python
orgs = arkova.list_orgs()
[org.display_name for org in orgs.organizations]
```

```python
arkova.close()
```
